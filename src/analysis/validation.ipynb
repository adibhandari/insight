```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
test = pd.read_csv('./data/test_set_all_a4.csv')
test.head()
```

```python
texas_dogs = (test[(test['contact.address.state']=='TX')])
#texas_dogs = (test[(test['contact.address.state']=='TX') & test['cg_adpt_time']=='> 3 months'])
texas_dogs.head()
```

```python
rand_texas_dogs = texas_dogs.sample(n=100, random_state=0)
```

```python
rand_texas_dogs.head()
```

```python
X_features = ['age', 'gender', 'size', 'breed_pop',
               'breeds.primary',
              'breeds.mixed',
             'contact.address.state']
rtd_X = rand_texas_dogs[X_features]
rtd_X = rtd_X.reset_index()
rtd_X.head()
```

```python
az = ['AZ'] * 100
co = ['CO'] * 100
mn = ['MN'] * 100
```

```python
import random
random.seed = 0
rand_state = random.choice(['AZ','MN','CO'])
rand_state
```

```python
rand_state = [random.choice(['AZ','MN','CO']) for i in range(100)]
```

```python
print('AZ :{}, CO: {}, MN: {}'.format(rand_state.count('AZ'), rand_state.count('CO'), rand_state.count('MN')))
```

```python
rtd_X['rand_state']=rand_state
rtd_X['state_az']=az
rtd_X['state_co']=co
rtd_X['state_mn']=mn
rtd_X.head()
```

```python
rtd_X.head()
```

```python
filename = './models/rfpipe_combined.pkl'
loaded_pipe = pickle.load(open(filename, 'rb'))
```

```python
rand_times=loaded_pipe.predict(rtd_X.drop(['index','contact.address.state', 'state_az', 'state_co', 'state_mn'], axis=1))
```

```python
rand_times=rand_times.tolist()
```

```python
def convert_time(time):
    if time == '< 1 week':
        return 3.5
    elif time == '1 - 2 weeks':
        return 10
    elif time == '< 1 month':
        return 21.5
    elif time == '< 3 months':
        return 59.5
    else:
        return 100
```

```python
rand_times
```

```python
rand_times_conv = list(map(lambda x:convert_time(x), rand_times))
rtd_X['adoption_time_rand']=rand_times_conv
```

```python
az_times=loaded_pipe.predict(rtd_X.drop(['index','contact.address.state', 'rand_state', 'state_co', 
                                         'state_mn', 'adoption_time_rand'], axis=1))
az_times=az_times.tolist()
az_times_conv = list(map(lambda x:convert_time(x), az_times))
```

```python
co_times=loaded_pipe.predict(rtd_X.drop(['index','contact.address.state', 'rand_state', 'state_az', 
                                         'state_mn', 'adoption_time_rand'], axis=1))
co_times=co_times.tolist()
co_times_conv = list(map(lambda x:convert_time(x), co_times))
```

```python
mn_times=loaded_pipe.predict(rtd_X.drop(['index','contact.address.state', 'rand_state', 'state_co', 
                                         'state_az', 'adoption_time_rand'], axis=1))
mn_times=mn_times.tolist()
mn_times_conv = list(map(lambda x:convert_time(x), mn_times))
```

```python
list(zip(az_times_conv, co_times_conv, mn_times_conv))
```

```python
min_time = [min(l1, l2, l3) for l1, l2, l3 in zip(az_times_conv, co_times_conv, mn_times_conv)]
```

```python
rtd_X['adoption_time_min']=min_time
rtd_X.head()
```
