```python
#merge data from several states
import numpy as np
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
state_info = pd.read_csv('stats.csv')
state_pos = 0
current_state = state_info['state'][state_pos]
current_city = state_info['city'][state_pos]
current_state
```

```python
alldogs_merged = pd.read_csv('alldogs_reduced_{}.csv'.format(current_state))
```

```python
state_pos=state_pos+1
```

```python
while(state_pos<7):
    current_state = state_info['state'][state_pos]
    current_city = state_info['city'][state_pos]
    temp = pd.read_csv('alldogs_reduced_{}.csv'.format(current_state))
    frames = [alldogs_merged, temp]
    alldogs_merged = pd.concat(frames)
    state_pos=state_pos+1
```

```python
export = alldogs_merged.to_csv('alldogs_merged.csv', header=True, index = False)
```
